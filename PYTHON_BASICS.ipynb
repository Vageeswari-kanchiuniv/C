```python
# PYTBASICS - Python Basics in Colab Notebook
# ===========================================

# 1. INTRODUCTION
print("Welcome to Python Basics üöÄ")

# -------------------------------------------
# 2. VARIABLES & DATA TYPES
x = 10
y = 3.14
name = "Alice"
is_ok = True

print("Integer:", x)
print("Float:", y)
print("String:", name)
print("Boolean:", is_ok)

# -------------------------------------------
# 3. INPUT & OUTPUT
# Uncomment to test input in Colab
# user_name = input("Enter your name: ")
# print("Hello,", user_name)

# -------------------------------------------
# 4. OPERATORS
a, b = 15, 4
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)

# -------------------------------------------
# 5. CONTROL FLOW
age = 20
if age >= 18:
    print("You are an adult.")
elif age > 12:
    print("You are a teenager.")
else:
    print("You are a child.")

# For Loop
for i in range(5):
    print("For Loop:", i)

# While Loop
n = 3
while n > 0:
    print("While Loop:", n)
    n -= 1

# -------------------------------------------
# 6. FUNCTIONS
def greet(name):
    return "Hello " + name

print(greet("Bob"))

# -------------------------------------------
# 7. DATA STRUCTURES
# List
fruits = ["apple", "banana", "cherry"]
print("List:", fruits)

# Tuple
point = (10, 20)
print("Tuple:", point)

# Set
nums = {1, 2, 3, 2}
print("Set:", nums)

# Dictionary
student = {"name": "John", "age": 20}
print("Dictionary:", student)

# -------------------------------------------
# 8. MODULES & PACKAGES
import math
print("Square root of 16:", math.sqrt(16))

# -------------------------------------------
# 9. FILE HANDLING
with open("sample.txt", "w") as f:
    f.write("Hello, Python!")

with open("sample.txt", "r") as f:
    print("File Content:", f.read())

# -------------------------------------------
# 10. EXCEPTION HANDLING
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Error: Cannot divide by zero ‚ùå")
finally:
    print("This block always runs ‚úÖ")
```
