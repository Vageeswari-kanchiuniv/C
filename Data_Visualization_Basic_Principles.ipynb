```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data: monthly sales
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [12000, 15000, 13000, 18000, 20000, 22000, 21000, 23000, 25000, 24000, 27000, 30000]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Set a clean style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Month', y='Sales', data=df, palette='Blues_d')

# Add clarity with labels and title
plt.title("Monthly Sales Overview (2025)", fontsize=16, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)

# Add value labels on top of bars for clarity
for i, value in enumerate(df['Sales']):
    plt.text(i, value + 500, f"${value:,}", ha='center', fontsize=9)

# Show the plot
plt.tight_layout()
plt.show()
```
