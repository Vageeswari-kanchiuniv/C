```python
# --- SETUP ---
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import HTML, Image, Video

# --- SAMPLE DATA ---
students = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
marks_test1 = np.random.randint(50, 90, size=10)
marks_test2 = marks_test1 + np.random.randint(-10, 15, size=10)
marks_test2 = np.clip(marks_test2, 0, 100)  # ensure marks are between 0100

# --- PLOT SETUP ---
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(students, marks_test1, color='skyblue')
ax.set_ylim(0, 100)
ax.set_title("Student Marks Animation")
ax.set_ylabel("Marks (%)")

# --- ANIMATION FUNCTION ---
def update(frame):
    # Interpolate marks between test1 and test2
    progress = frame / 50  # total frames = 50
    current_marks = marks_test1 + (marks_test2 - marks_test1) * progress

    for bar, mark in zip(bars, current_marks):
        bar.set_height(mark)
    ax.set_title(f"Student Marks Progress (Frame {frame})")
    return bars

# --- CREATE ANIMATION ---
ani = FuncAnimation(fig, update, frames=51, interval=100, blit=False)

# --- DISPLAY INLINE IN COLAB ---
HTML(ani.to_jshtml())
```

```python
gif_writer = PillowWriter(fps=20)
ani.save("student_marks.gif", writer=gif_writer)
print(" GIF saved as student_marks.gif")

Image(filename="student_marks.gif")
```

```python
ani.save("student_marks.mp4", writer="ffmpeg", fps=20)
print(" Video saved as student_marks.mp4")

Video("student_marks.mp4")
```

```python
# --- SETUP ---
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import HTML, Image, Video

# --- SAMPLE DATA ---
students = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# Random marks for demonstration
np.random.seed(42)
marks_test1 = np.random.randint(50, 90, size=10)
marks_test2 = marks_test1 + np.random.randint(-15, 20, size=10)
marks_test2 = np.clip(marks_test2, 0, 100)

# Identify which students improved
improved = marks_test2 > marks_test1

# --- PLOT SETUP ---
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(students, marks_test1, color=['skyblue' if imp else 'lightcoral' for imp in improved])
ax.set_ylim(0, 100)
ax.set_title("Student Marks: Test 1 to Test 2")
ax.set_ylabel("Marks (%)")

# Add a legend manually
handles = [
    plt.Rectangle((0, 0), 1, 1, color='skyblue', label='Improved'),
    plt.Rectangle((0, 0), 1, 1, color='lightcoral', label='Declined')
]
ax.legend(handles=handles, loc='upper left')

# --- ANIMATION FUNCTION ---
def update(frame):
    progress = frame / 50  # progress between 0 and 1
    current_marks = marks_test1 + (marks_test2 - marks_test1) * progress

    for bar, mark, imp in zip(bars, current_marks, improved):
        bar.set_height(mark)
        # Keep color constant (blue if improved, red if declined)
        bar.set_color('skyblue' if imp else 'lightcoral')

    ax.set_title(f"Student Marks Transition (Frame {frame})")
    return bars

# --- CREATE ANIMATION ---
ani = FuncAnimation(fig, update, frames=51, interval=100, blit=False)

# --- DISPLAY INLINE ---
HTML(ani.to_jshtml())
```

```python
gif_writer = PillowWriter(fps=20)
ani.save("student_marks_colored.gif", writer=gif_writer)
print(" GIF saved as student_marks_colored.gif")

Image(filename="student_marks_colored.gif")
```

```python
ani.save("student_marks_colored.mp4", writer="ffmpeg", fps=20)
print(" Video saved as student_marks_colored.mp4")

Video("student_marks_colored.mp4")
```
