# ðŸ“˜ Data Analysis with Python
Demonstration of **Data Loading, Data Wrangling, Data Exploration, Aggregation, and Visualization** using Pandas, NumPy, Matplotlib, and Seaborn.

## ðŸ”¹ 1. Data Loading
We will create a **sample employee dataset** inside the program (no external file required).

```python

import pandas as pd
import numpy as np

# Sample dataset
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 60000, 75000, 80000, np.nan, 55000, 70000, 90000],
    "Age": [25, 30, 35, 40, 29, np.nan, 32, 45],
    "Joining_Year": [2018, 2019, 2020, 2018, 2019, 2021, 2020, 2017]
}

df = pd.DataFrame(data)
df
```

## ðŸ”¹ 2. Data Wrangling
We will clean missing values and create a new feature (`Experience`).

```python

# Fill missing values
df["Salary"].fillna(df["Salary"].mean(), inplace=True)
df["Age"].fillna(df["Age"].median(), inplace=True)

# Feature Engineering
df["Experience"] = 2025 - df["Joining_Year"]

df
```

## ðŸ”¹ 3. Data Exploration (EDA)
Check summary statistics, data types, and correlations.

```python

print("Summary Statistics:\n", df.describe())
print("\nData Types:\n", df.dtypes)
print("\nCorrelation:\n", df.corr(numeric_only=True))
```

## ðŸ”¹ 4. Data Aggregation & Group Operations
Compute **average salary by department**.

```python

grouped = df.groupby("Department")["Salary"].mean()
grouped
```

## ðŸ”¹ 5. Plotting & Visualization
We will visualize the dataset using Matplotlib and Seaborn.

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Bar Plot - Average Salary by Department
plt.figure(figsize=(6,4))
sns.barplot(x=grouped.index, y=grouped.values, palette="viridis")
plt.title("Average Salary by Department")
plt.ylabel("Salary")
plt.xlabel("Department")
plt.show()
```

```python

# Histogram - Age distribution
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=5, kde=True, color="blue")
plt.title("Age Distribution of Employees")
plt.show()
```

```python

# Scatter Plot - Age vs Salary
plt.figure(figsize=(6,4))
sns.scatterplot(x="Age", y="Salary", data=df, hue="Department", style="Department", s=100)
plt.title("Age vs Salary by Department")
plt.show()
```

```python

# Boxplot - Salary distribution by Department
plt.figure(figsize=(6,4))
sns.boxplot(x="Department", y="Salary", data=df, palette="Set2")
plt.title("Salary Distribution by Department")
plt.show()
```
